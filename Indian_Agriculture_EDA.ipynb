###### Exploratory Data Analysis (EDA) refers to the process of studying and exploring data sets to comprehend their dominant traits, discover patterns, locate outliers, and identify relationships between variables.EDA is normally carried out as a preliminary step before undertaking extra formal statistical analyses or modeling.

###### Below is EDA of India's Agricultual land usage data from 1966 to 2017 across 20 states. Real time issue faced is data is huge and also to cover all aspects in single chart or table is difficult.At certain stages data is split netween more than one chart or table.Data set has 16146 rows and 80 columns

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df = pd.read_csv('ICRISAT-District Level Data.csv')
df.head(5)
```

```python
df2 = df.groupby(['State Name','Year'])[['RICE AREA (1000 ha)','WHEAT AREA (1000 ha)','SORGHUM AREA (1000 ha)',
       'KHARIF SORGHUM AREA (1000 ha)','RABI SORGHUM AREA (1000 ha)','SORGHUM AREA (1000 ha)',
       'PEARL MILLET AREA (1000 ha)','MAIZE AREA (1000 ha)','FINGER MILLET AREA (1000 ha)',
       'BARLEY AREA (1000 ha)','CHICKPEA AREA (1000 ha)','PIGEONPEA AREA (1000 ha)',
       'MINOR PULSES AREA (1000 ha)','GROUNDNUT AREA (1000 ha)','SESAMUM AREA (1000 ha)',
       'RAPESEED AND MUSTARD AREA (1000 ha)','SAFFLOWER AREA (1000 ha)','CASTOR AREA (1000 ha)',
       'LINSEED AREA (1000 ha)','SUNFLOWER AREA (1000 ha)','SOYABEAN AREA (1000 ha)','OILSEEDS AREA (1000 ha)',
       'SUGARCANE AREA (1000 ha)','COTTON AREA (1000 ha)','FRUITS AREA (1000 ha)','VEGETABLES AREA (1000 ha)',
       'POTATOES AREA (1000 ha)','ONION AREA (1000 ha)','FODDER AREA (1000 ha)']].sum()
df3 = df2.reset_index()
df4 = df3.drop(columns=['State Name'],axis=1)
df5 = df4.set_index('Year')
df6 = df5.loc[2017].sum(axis=0)
df7 = pd.DataFrame(df6).reset_index().rename(columns={'index':'Crop',0:'Area Sown'})
df7['Area Sown'] = round(df7['Area Sown']/1000)
df7['Crop']=df7['Crop'].apply(lambda x:x.split('AREA (1000 ha)')[0])
df7.head()
```

###### As per given data and from below chart in Fig 1 it is seen that Rice, Wheat, Chickpea,Minor Pulses, Soyabean and Cotton put together take major land area sown across states. A point call out that below few charts will show trends for these crops due to limitaion to display 

```python
plt.figure(figsize=(8, 6))
x = list(df7['Area Sown'])
y = list(df7['Crop'])
plt.barh(y,x,color='orange')
plt.title('Fig 1 : Crop wise total area under cultivation in the year 2017')
plt.xlabel('Area under Cultivation(million hectare)')
plt.ylabel('Crops')
#plt.tight_layout()
```

###### Below line graph in Fig 2 shows land usage patterns for cultivation for all states put together in India from 1966 to 2017(40 Years)

```python
df8 = pd.DataFrame(df5.sum(axis=1)).rename(columns={0:'Area'}).reset_index()
df9 = df8.groupby('Year')['Area'].sum()
df10 = pd.DataFrame(df9).reset_index()
#df10
plt.figure(figsize=(10, 4))
x = df10['Year']
y = df10['Area']/1000
plt.plot(x,y,color='green',marker='.')
plt.title('Fig 2 : Land Area Use in Last 40 years for Cultivation')
plt.xlabel('Year')
plt.ylabel('Area(in million hectares)')
#plt.tight_layout()
plt.show()
```

```python
df12 = df3.drop(columns=['Year'],axis=1).groupby('State Name')[[
       'RICE AREA (1000 ha)','WHEAT AREA (1000 ha)','MAIZE AREA (1000 ha)',
       'CHICKPEA AREA (1000 ha)','MINOR PULSES AREA (1000 ha)',
       'SOYABEAN AREA (1000 ha)','COTTON AREA (1000 ha)',
       ]].sum()
df12.head()
```

```python
df13 = df12.drop('Himachal Pradesh') #Due to negative values for few columns
df13 = df13.loc[:].div(1000).round(2)
df13.columns = df13.columns.str.replace('AREA (1000 ha)','')
df13
```

###### Below chart shows crops grows over different states. Thses are few crops which seemed to be grown over karger area

```python
df_AB = df13.reset_index()
x1 = df_AB['State Name']
y1 = df_AB['COTTON ']
y2 = df_AB['WHEAT ']
y3 = df_AB['MAIZE ']
y4 = df_AB['MINOR PULSES ']
y5 = df_AB['CHICKPEA ']
y6 = df_AB['SOYABEAN ']
y7 = df_AB['RICE ']
fig = plt.figure(figsize=(8, 6))
plt.bar(x1, y1,color='red')
plt.bar(x1, y2,color='green')
plt.bar(x1, y3,color='black')
plt.bar(x1, y4,color='#8c2ac3')
plt.bar(x1, y5,color='blue')
plt.bar(x1, y6,color='orange')
plt.bar(x1, y7,color='#cda7bb')
plt.ylabel('Area Sown(million hectares)')
plt.xticks(rotation=90)
plt.legend(['COTTON ','WHEAT ','MAIZE ','MINOR PULSES ','CHICKPEA ','SOYABEAN ','RICE '])
plt.title('Fig 3 : Across States Area Cultivated')
plt.show()
```

###### Below pie chart shows share of cultivated area for different crops in Gulbarga/Kalaburagu district of Karnataka State 

```python
df21 = df.copy().reset_index(drop=True)
df21.head(5)
df22 = df21.groupby(['State Name'])
df23 = df22.get_group('Karnataka')
df24 = df23.groupby('Dist Name')[['RICE AREA (1000 ha)','WHEAT AREA (1000 ha)',
       'SORGHUM AREA (1000 ha)',
       'PEARL MILLET AREA (1000 ha)',
       'BARLEY AREA (1000 ha)','CHICKPEA AREA (1000 ha)','PIGEONPEA AREA (1000 ha)',
       'MINOR PULSES AREA (1000 ha)','GROUNDNUT AREA (1000 ha)',
       'SAFFLOWER AREA (1000 ha)',
       'SUNFLOWER AREA (1000 ha)','SOYABEAN AREA (1000 ha)','OILSEEDS AREA (1000 ha)',
       'COTTON AREA (1000 ha)','FRUITS AREA (1000 ha)',
       'POTATOES AREA (1000 ha)','FODDER AREA (1000 ha)']].sum()

df24.columns = df24.columns.str.replace('AREA (1000 ha)','')
df24 = df24.loc[:].div(1000).round(2)
df25 = df24.reset_index()
ftr = df25['Dist Name'] == 'Gulbarga / Kalaburagi'
df_kalaburagi = df25.loc[ftr]
df_kalaburagi = df_kalaburagi.drop(['BARLEY ','SOYABEAN ','POTATOES ','FODDER '],axis=1)
df_kalbrg = df_kalaburagi.loc[:,~df_kalaburagi.columns.duplicated()].copy()
df_kalbrg = df_kalbrg.drop('Dist Name',axis=1)
t = df_kalbrg.transpose().reset_index().rename(columns={'index':'Crop',9:'Area'})
x = list(t['Area'])
l = list(t['Crop'])
fig = plt.figure(figsize=(10, 7))
plt.pie(x,labels=l,autopct='%1.1f%%')
plt.legend(title="Crops",loc="center left",bbox_to_anchor=(1.5, 0, 0.5, 1))
plt.title('Fig4 : Land used(in %) for sowing in Gulbarga / Kalaburagi District of Karnataka State')
plt.show()
```

```python
dff = df3.copy()
dff.columns = dff.columns.str.replace('AREA (1000 ha)','')
dff =  dff.set_index(['State Name','Year'])
dff =  dff.loc[:].div(1000).round(2)
dff_sum = pd.DataFrame(dff.sum(axis=1)).reset_index()
dff_sum1 = dff_sum.rename(columns={0:'Area'})
dff_sum2 = dff_sum1[dff_sum1['Year'] >= 1997]
dff_sum3 = dff_sum2.pivot(index='State Name',columns='Year',values='Area')
dff_sum3
plt.figure(figsize=(10,8))
sns.heatmap(dff_sum3,cmap='Blues',annot=True)
plt.title('Fig5 :Land used across states for sowing')
plt.show()
```

```python
df_sss = df[['State Name','RICE AREA (1000 ha)','RICE YIELD (Kg per ha)','WHEAT AREA (1000 ha)','WHEAT YIELD (Kg per ha)',
'MINOR PULSES AREA (1000 ha)','MINOR PULSES YIELD (Kg per ha)','CHICKPEA AREA (1000 ha)','CHICKPEA YIELD (Kg per ha)',
'COTTON AREA (1000 ha)','COTTON YIELD (Kg per ha)']]

dff_agg = df_sss.groupby('State Name').sum().loc[:].div(1000).round(2)
dff_agg.columns = dff_agg.columns.str.replace('AREA (1000 ha)','')
dff_agg.columns = dff_agg.columns.str.replace('(Kg per ha)','')
dff_agg.reset_index()
dff_agg1 = dff_agg.loc['Andhra Pradesh':'Kerala']

r = list(dff_agg1['RICE '])
ry = list(dff_agg1['RICE YIELD '])
w = list(dff_agg1['WHEAT '])
wy = list(dff_agg1['WHEAT YIELD '])
mp = list(dff_agg1['MINOR PULSES '])
mpy = list(dff_agg1['MINOR PULSES YIELD '])
cp = list(dff_agg1['CHICKPEA '])
cpy = list(dff_agg1['CHICKPEA YIELD '])
ct = list(dff_agg1['COTTON '])
cty = list(dff_agg1['COTTON YIELD '])
x = np.arange(len(dff_agg1.index.tolist())) 
x_indexes=np.arange(len(dff_agg1.index.tolist()))

width=0.1
plt.style.use('fivethirtyeight')
plt.figure(figsize=(17,6))
plt.bar(x_indexes-width-width-width,r,width=width,color='red',label='Rice')
plt.bar(x_indexes-width-width,ry,width=width,color='green',label='Rice Yield')
plt.bar(x_indexes-width,w,width=width,color='blue',label='Wheat')
plt.bar(x_indexes,wy,width=width,color='purple',label='Wheat Yield')
plt.bar(x_indexes+width,mp,width=width,color='black',label='Minor Pulses')
plt.bar(x_indexes+width+width,mpy,width=width,color='orange',label='Minor Pulses Yield')
plt.bar(x_indexes+width+width+width,cp,width=width,color='brown',label='Chickpea')
plt.bar(x_indexes+width+width+width,cpy,width=width,color='#0099ff',label='Chickpea Yield')
plt.bar(x_indexes+width+width+width,ct,width=width,color='#4d5d53',label='Cotton')
plt.bar(x_indexes+width+width+width,cty,width=width,color='#d2691e',label='Cotton Yield')
plt.xticks(x, dff_agg1.index.tolist(),rotation=90)
plt.title('Fig6.1 : Across States Crop Area Sown Vs Crop Yield')
plt.xlabel('State Name')
plt.ylabel('Crop Area Sown-Crop Yield')
plt.legend()
plt.show()
```

```python
dff_agg2 = dff_agg.loc['Madhya Pradesh':'West Bengal']
r1 = list(dff_agg2['RICE '])
ry1 = list(dff_agg2['RICE YIELD '])
w1 = list(dff_agg2['WHEAT '])
wy1 = list(dff_agg2['WHEAT YIELD '])
mp1 = list(dff_agg1['MINOR PULSES '])
mpy1 = list(dff_agg2['MINOR PULSES YIELD '])
cp1 = list(dff_agg2['CHICKPEA '])
cpy1 = list(dff_agg2['CHICKPEA YIELD '])
ct1 = list(dff_agg2['COTTON '])
cty1 = list(dff_agg2['COTTON YIELD '])
x = np.arange(len(dff_agg2.index.tolist())) 
x_indexes=np.arange(len(dff_agg2.index.tolist()))

width=0.1
plt.style.use('fivethirtyeight')
plt.figure(figsize=(25,6))
plt.bar(x_indexes-width-width-width,r1,width=width,color='red',label='Rice')
plt.bar(x_indexes-width-width,ry1,width=width,color='green',label='Rice Yield')
plt.bar(x_indexes-width,w1,width=width,color='blue',label='Wheat')
plt.bar(x_indexes,wy1,width=width,color='purple',label='Wheat Yield')
plt.bar(x_indexes+width,mp1,width=width,color='black',label='Minor Pulses')
plt.bar(x_indexes+width+width,mpy1,width=width,color='orange',label='Minor Pulses Yield')
plt.bar(x_indexes+width+width+width,cp1,width=width,color='brown',label='Chickpea')
plt.bar(x_indexes+width+width+width,cpy1,width=width,color='#0099ff',label='Chickpea Yield')
plt.bar(x_indexes+width+width+width,ct1,width=width,color='#4d5d53',label='Cotton')
plt.bar(x_indexes+width+width+width,cty1,width=width,color='#d2691e',label='Cotton Yield')
plt.xticks(x, dff_agg2.index.tolist(),rotation=90)
plt.title('Fig6.2 : Across States Crop Area Sown Vs Crop Yield')
plt.xlabel('State Name')
plt.ylabel('Crop Area Sown-Crop Yield')
plt.legend()
plt.show()
```

```python
dff_pp = df[['State Name','RICE PRODUCTION (1000 tons)','WHEAT PRODUCTION (1000 tons)','MINOR PULSES PRODUCTION (1000 tons)']]
dff_pp1 = dff_pp.groupby(['State Name']).sum().loc[:].div(1000).round(2)
dff_pp1.columns = dff_pp1.columns.str.replace('UCTION (1000 tons)','')
dff_pp2 = dff_pp1
df_pp3 = dff_pp2.reset_index()
x = list(df_pp3['State Name'])
y1 = list(df_pp3['RICE PROD'])
y2 = list(df_pp3['WHEAT PROD'])
y3 = list(df_pp3['MINOR PULSES PROD'])
x2 = np.arange(len(df_pp3['State Name'].tolist())) 
plt.figure(figsize=(10, 4))
plt.plot(x,y1,color='green',marker='.',label='Rice Production')
plt.plot(x,y2,color='brown',marker='.',label='Wheat Production')
plt.plot(x,y3,color='orange',marker='.',label='Minor Pulses Production')
plt.xticks(x2, df_pp3['State Name'].tolist(),rotation=90)
plt.title('Fig 7 :Rice, Wheat and Minor Pluses Production across states(million tons/hectare)')
plt.legend()
plt.show()
```
